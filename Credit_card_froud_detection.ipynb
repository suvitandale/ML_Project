```python
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
new_credit = credit_card_data.dropna(axis=0)
```

```python
new_credit.Class.value_counts()
```

```python
legit = new_credit[new_credit['Class']==0].sample(222,axis=0)
froud = new_credit[new_credit['Class']==1]
```

```python
updated_credit = pd.concat([legit,froud],axis=0)
```

```python
updated_credit.groupby('Class').mean()
```

```python
legit.shape
```

```python
froud.shape
```

##Split data into Train and Test Datasets

```python
X = updated_credit.drop(columns='Class')
```

```python
Y = updated_credit['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=42)
```

# Model Building

```python
model = LogisticRegression()
```

```python
model.fit(train_x,train_y)
```

```python
predict_x_train = model.predict(train_x)
```

```python
acrcy = accuracy_score(train_y,predict_x_train)
print(acrcy)
```

```python
predict_y = model.predict(test_x)
```

```python
acrcy_y = accuracy_score(test_y,predict_y)
print(acrcy_y)
```
